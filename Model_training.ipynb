```python
from ultralytics import YOLO

# Load the model
model = YOLO('yolov8n-seg.pt')  # Choose the model variant as needed

# Train the model with specified parameters including patience
model.train(
    data='/kaggle/input/teeth-assignment-data/teeth_dataset_afterlook/data.yaml',  # Path to your data.yaml file
    epochs=150,                # Number of epochs to train
    project='training_results',  # Directory to save results
    name='experiment01',       # Name of the experiment (subdirectory)
    save_period=15,            # Save checkpoint every 15 epochs
    batch=8,                   # Batch size for training
    workers=4,                 # Number of data loader workers
    patience=20                # Number of epochs to wait for improvement before stopping
)
```
